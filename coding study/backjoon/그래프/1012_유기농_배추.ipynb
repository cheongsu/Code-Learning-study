```python
# 초기 코드

A,B,N = map(int,input().split())

C = max(A,B)


visited = [-1]*(C+1)
graph = [[] for i in range(C+1)]
for _ in range(N):
    a,b = map(int,input().split())
    graph[a].append(b)
    graph[b].append(a)
    visited[a] = -2


for i in range(1,len(graph)):
    graph[i].sort(reverse=True)
print(graph)
def DFS(start):
    visited[start] = -1
    
    for adj_node in graph[start]:
        if visited[adj_node] == -2:
            DFS(adj_node)
    return 
count = 1
for k in range(len(graph)):
    DFS(k)
    if visited in -2:
        DFS(-1)
        count +=1
        
```
