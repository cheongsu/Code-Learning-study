문제
지민이는 N개의 원소를 포함하고 있는 양방향 순환 큐를 가지고 있다. 지민이는 이 큐에서 몇 개의 원소를 뽑아내려고 한다.

지민이는 이 큐에서 다음과 같은 3가지 연산을 수행할 수 있다.

첫 번째 원소를 뽑아낸다. 이 연산을 수행하면, 원래 큐의 원소가 a1, ..., ak이었던 것이 a2, ..., ak와 같이 된다.
왼쪽으로 한 칸 이동시킨다. 이 연산을 수행하면, a1, ..., ak가 a2, ..., ak, a1이 된다.
오른쪽으로 한 칸 이동시킨다. 이 연산을 수행하면, a1, ..., ak가 ak, a1, ..., ak-1이 된다.
큐에 처음에 포함되어 있던 수 N이 주어진다. 그리고 지민이가 뽑아내려고 하는 원소의 위치가 주어진다. (이 위치는 가장 처음 큐에서의 위치이다.) 이때, 그 원소를 주어진 순서대로 뽑아내는데 드는 2번, 3번 연산의 최솟값을 출력하는 프로그램을 작성하시오.

입력
첫째 줄에 큐의 크기 N과 뽑아내려고 하는 수의 개수 M이 주어진다. N은 50보다 작거나 같은 자연수이고, M은 N보다 작거나 같은 자연수이다. 둘째 줄에는 지민이가 뽑아내려고 하는 수의 위치가 순서대로 주어진다. 위치는 1보다 크거나 같고, N보다 작거나 같은 자연수이다.

현재 위치에서 index를 구한다.

1번의 경우 2는 index 1이 나옴

len(list)-index를 하여 len(list)/2 보다 크면 index만큼 rotate하고
반대면 반대방향으로 회전해야 하므로 반대방향으로 len(list)-index만큼 rotate한다.
로테이트한 숫자를 count에 더해서 저장한다.

```python
from collections import deque
n, m = map(int,input().split())

my_list = list(map(int,input().split()))
queue = []

def rotate1(arr, x):
    if x==0:
        return arr
    else:
        left = arr[:-n]
        right = arr[-n:]
        return left + right

def rotate2(arr, x):
    if x==0:
        return arr
    else:
        right = arr[:-n]
        left = arr[-n:]
        return left + right

for i in range(n):
    queue.append(i+1)

count = 0
for i in my_list:
    print(queue)
    print(queue.index(i))
    if len(queue) - queue.index(i) > len(queue)/2:
        rotate2(queue,queue.index(i))
        count +=queue.index(i)
        print('count',count)
        del queue[0]
    else:
        rotate1(queue,len(queue) - queue.index(i))
        count +=len(queue) - queue.index(i)
        del queue[0]
        print('count',count)
print(queue)
print(count)

# deque는 index를 구하는 함수가 없다.

#rotate 함수가 작동하지 않음
```

```python
from collections import deque
n, m = map(int,input().split())

position = list(map(int,input().split()))
dq = deque([i for i in range(1, n+1)])



count = 0   
for i in position:  
    while True:     
        if dq[0] == i:  
            dq.popleft()
            break
        else:
            if dq.index(i) < len(dq)/2:  
                while dq[0] != i:   # dq의 첫번째 인덱스가 i와 같아질때까지 반복
                    dq.append(dq.popleft())  
                    count += 1
            else:   # 뽑아내려는 수의 위치 인덱스가 dq의 길이를 반으로 나눈것보다 클때는 오른쪽으로 움직여야 최소
                while dq[0] != i:
                    dq.appendleft(dq.pop())  
                    count += 1
print(count)


# rotate 후 제거가 아닌, pop()을 하면서 동시에 반대편 위치에 append를 하는 아이디어 적용
```
